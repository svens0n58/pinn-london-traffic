```python
import pandas as pd
import folium
```

```python
"""file_path = 'Singapore_cameras_new.xlsx'  # Change to your actual path
df = pd.read_excel(file_path)"""
file_path = 'traffic_data/detectors_public.csv'  # Change to your actual path
df = pd.read_csv(file_path)
df
```

```python
# Load the Excel file

# Create a base map centered around the mean latitude and longitude values
camera_map = folium.Map(location=[df['lat'].mean(), df['long'].mean()], zoom_start=12)

"""for _, row in df.iterrows():
    if row['2022_Direction'] == 0:
        marker_color = 'blue'
    elif row['2022_Direction'] == 1:
        marker_color = 'green'
    elif row['2022_Direction'] == 2:
        marker_color = 'red'
    else:
        marker_color = 'gray'

    folium.Marker(
        location=[row['lat'], row['long']],
        popup=f"Camera at ({row['lat']}, {row['long']}, Camera_ID: {row['detid']})",
        icon=folium.Icon(color=marker_color)
    ).add_to(camera_map)"""
city_name = 'london'
for _, row in df.iterrows():
    if row['citycode'] == city_name:
        folium.Marker(
            location=[row['lat'], row['long']],
            popup=f"Camera at ({row['lat']}, {row['long']}, Lanes: {row['lanes']}, Camera_ID: {row['detid']})",
            icon=folium.Icon()
        ).add_to(camera_map)


# Save the map to an HTML file
camera_map.save("traffic_data/maps/camera_map_" + city_name + ".html")

print("Map has been saved to 'camera_map_" + city_name + ".html'")
```
